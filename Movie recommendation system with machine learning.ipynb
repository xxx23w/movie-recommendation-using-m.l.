# Data Preprocessing

```python

#Rename column
credits_column_renamed = credits.rename(index = str, columns = {"movie_id": "id"})
#Merge column data
movie_df_merge = movie_df.merge(credits_column_renamed, on='id')
#check data
movie_df_merge.head()
```

```python
#drop unneccessary columns
movie_cleaned_df = movie_df_merge.drop(columns=['homepage', 'title_x', 'title_y', 'status','production_countries'])
print(movie_cleaned_df.head()) #check data
print(movie_cleaned_df.info())#check data info
print(movie_cleaned_df.head(1)['overview'])#check data
```

# APPly tf-idf vectorizer

```python
#import library
from sklearn.feature_extraction.text import TfidfVectorizer

#Apply vectorizer
tfv = TfidfVectorizer(min_df = 3, max_features = None,
                    strip_accents = 'unicode', analyzer = 'word', token_pattern = r'\w{1,}',#remove unnecessary character
                    ngram_range = (1, 3), #Take 1 to 3 combinations of word
                     stop_words = 'english') #remove unneccessary words
                    
 #fill NaNs with empty string
movie_cleaned_df['overview'] = movie_cleaned_df['overview'].fillna('')
```

```python
#fit the tf-idf on 'overview' text
tfv_matrix = tfv.fit_transform(movie_cleaned_df['overview'])
#sparse matrix
tfv_matrix
```

```python
tfv_matrix.shape
```

```python
#import library
from sklearn.metrics.pairwise import sigmoid_kernel

#compute the sigmoid kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
#Reverse mapping of indices and movie titles
indices = pd.Series(movie_cleaned_df.index, index = movie_cleaned_df['original_title']).drop_duplicates()
```

```python
# Recommender function
def give_rec(title, sig = sig):
    
    # Get the index corresponding to original_title
    idx = indices[title]

    # Get the pairwsie similarity scores 
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies 
    sig_scores = sorted(sig_scores, key = lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies
    sig_scores = sig_scores[1:11]

    # Movie indices
    movie_indices = [i[0] for i in sig_scores]

    # Top 10 most similar movies
    return movie_cleaned_df['original_title'].iloc[movie_indices]
```

```python
#result
# Test content-based recommendation system with the seminal film The Dark Knight Rises
give_rec('The Dark Knight Rises')
```
